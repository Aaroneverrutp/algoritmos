```python
### desafio 2:

#crea un arbol  con raiz 20
#hijo iz 10 u derecho 30

class Nodo:
    def __init__(self,valor):
        self.valor= valor
        self.iz = None
        self.der = None

raiz = Nodo(20)
raiz.iz = Nodo(10)
raiz.der = Nodo(30)

print (f"el valor del nodo es: {raiz.valor}")
print (f"el valor del nodo izquierdo es: {raiz.iz.valor}")
print (f"el valor del nodo derecho es : {raiz.der.valor}")
```

```python
##Desafio 3:

# III . insertar valores

#regla:
# si el valor es menor -> ir a la izquierda
# si el valor es mayor -> ir a la derecha
# repetir hasta encontrar un espacio vacio


def insertar(raiz,valor):
    if raiz is None:
        return  Nodo(valor)
    # si el valor es menor
    if valor < raiz.valor:
        raiz.iz = insertar(raiz.iz, valor)
    else:
        raiz.der = insertar(raiz.der,valor)

    return raiz


#ejemplo construir este arbol :

raiz = None
raiz = insertar(raiz,10)
raiz = insertar(raiz,5)
raiz = insertar(raiz,15)
raiz = insertar(raiz,3)


print(f"raiz:{raiz.valor}")

print(f"izquierda nodo:{raiz.iz.valor}")
print(f"derecha nodo:{raiz.der.valor}")
print(f"izquierda de izquierda:{raiz.iz.iz.valor}")
```

```python
#IV. Buscar Valor
#Regla:

"""
-Si encuentras el valor -Exito
-Si es menos -> busca a la izquierda
-Si es mayor -> busca en la derecha
-Si llegas a nane -> no existe
"""

def buscar(raiz, valor):
    #Caso 1: Arbol vacio o no encontrado
    if raiz is None:
        return False
    #Cas2: Cuando lo encontramos
    if raiz.valor == valor:
        return True
    #Caso 3: Buscar izquierda
    if valor < raiz.valor:
        return buscar(raiz.iz, valor)
    #Caso 4: Buscar derecha
    return buscar(raiz.der, valor)

#Crear Arbol: 10,5,15,3,7
raiz = None
for valor in [10,5,15,3,7]:
    raiz = insertar(raiz, valor)

#Buscar valores:
print(f"¿Estara el 7?: {buscar(raiz, 7)}")
print(f"¿Estara el 20?: {buscar(raiz, 20)}")
```
