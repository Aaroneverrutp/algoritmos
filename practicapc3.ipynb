```python

# Implementación de árboles binarios
class Nodo:
    # Cada nodo tiene un valor y 2 hijos: izquierdo y derecho
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None

# Función para insertar nodos
def insertar(raiz, valor):
    """
    Inserta un nuevo valor en el árbol binario.
    - Si el valor es menor, va al subárbol izquierdo.
    - Si el valor es mayor, va al subárbol derecho.
    * Representación de un árbol binario de búsqueda.
    """
    # Si el árbol está vacío, crear un nuevo nodo
    if raiz is None:
        return Nodo(valor)
    # Si el valor es menor, insertar en el subárbol izquierdo
    if valor < raiz.valor:
        raiz.izquierdo = insertar(raiz.izquierdo, valor)
    # Si el valor es mayor, insertar en el subárbol derecho
    if valor > raiz.valor:
        raiz.derecho = insertar(raiz.derecho, valor)
    # Devolvemos la raíz sin cambios
    return raiz

# Función para evaluar la expresión del árbol
def evaluar(raiz):
    """
    Evalúa el árbol de expresión.
    Si el nodo es un número, devuelve su valor.
    Si el nodo es un operador, evalúa los subárboles izquierdo y derecho y aplica el operador.
    """
    if raiz is None:
        return 0
    
    # Si el nodo es un número, devolver su valor
    if isinstance(raiz.valor, int):
        return raiz.valor
    
    # Si el nodo es un operador, aplicar recursión en los subárboles izquierdo y derecho
    izquierdo = evaluar(raiz.izquierdo)
    derecho = evaluar(raiz.derecho)
    
    if raiz.valor == '+':
        return izquierdo + derecho
    elif raiz.valor == '-':
        return izquierdo - derecho
    elif raiz.valor == '*':
        return izquierdo * derecho
    elif raiz.valor == '/':
        return izquierdo / derecho
    else:
        raise ValueError(f"Operador desconocido: {raiz.valor}")

# Recorridos del árbol
def recorrido_inorden(raiz):
    """
    Recorrido IN-ORDEN: Izquierdo -> Raíz -> Derecho
    Este recorrido muestra los valores en orden "ascendente".
    """
    if raiz:
        recorrido_inorden(raiz.izquierdo)
        print(raiz.valor, end=" ")
        recorrido_inorden(raiz.derecho)
        
def recorrido_preorden(raiz):
    """
    Recorrido Pre-orden: Raíz -> Izquierdo -> Derecho
    Util para copiar o clonar un árbol.
    """
    if raiz:
        print(raiz.valor, end=" ")
        recorrido_preorden(raiz.izquierdo)
        recorrido_preorden(raiz.derecho)

def recorrido_postorden(raiz):
    """
    Recorrido Post-orden: Izquierdo -> Derecho -> Raíz
    Util para eliminar el árbol o evaluar expresiones.
    """
    if raiz:
        recorrido_postorden(raiz.izquierdo)
        recorrido_postorden(raiz.derecho)
        print(raiz.valor, end=" ")

# Crear el árbol de expresión
raiz = Nodo('*')
raiz.izquierdo = Nodo('+')
raiz.derecho = Nodo('-')
raiz.izquierdo.izquierdo = Nodo(7)
raiz.izquierdo.derecho = Nodo(3)
raiz.derecho.izquierdo = Nodo(5)
raiz.derecho.derecho = Nodo(2)

# Mostrar recorrido Pre-orden, In-orden y Post-orden
print("=== RECORRIDOS ===")
print("Preorden:")
recorrido_preorden(raiz)
print()
print("Inorden:")
recorrido_inorden(raiz)
print()
print("Postorden:")
recorrido_postorden(raiz)
print()

# Evaluar la expresión
resultado = evaluar(raiz)

# Mostrar la expresión y el resultado
print("\n=== ÁRBOL DE EXPRESIÓN ===")
print(f"Expresión: (7 + 3) * (5 - 2)")
print(f"Resultado: {resultado}")
```
