```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None #El hijo de la izquierda
        self.derecho = None #EL hijo de la derecha

#Crear arbol
raiz= Nodo("A")
raiz.izquierdo = Nodo("B")
raiz.derecho = Nodo("C")
raiz.izquierdo.izquierdo = Nodo("D")
raiz.izquierdo.derecho = Nodo("E")
raiz.izquierdo.izquierdo.izquierdo = Nodo("H")
raiz.izquierdo.izquierdo.derecho = Nodo("I")
raiz.izquierdo.derecho.derecho = Nodo("K")
raiz.derecho.izquierdo = Nodo("F")
raiz.derecho.derecho = Nodo("G")
raiz.derecho.izquierdo.izquierdo = Nodo("L")
raiz.derecho.izquierdo.derecho = Nodo("M")



#Recorrer arbol
def recorrer(nodo):
    if nodo:
        recorrer(nodo.izquierdo)
        print(nodo.valor)
        recorrer(nodo.derecho)

#Prueba:
recorrer(raiz)
```

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None #El hijo de la izquierda
        self.derecho = None #EL hijo de la derecha


raiz = Nodo(".")
raiz.izquierdo = Nodo("T")
raiz.izquierdo.izquierdo = Nodo("A")
raiz.izquierdo.derecho = Nodo("A")
raiz.izquierdo.izquierdo.izquierdo = Nodo("K")

raiz.derecho = Nodo("U")
raiz.derecho.izquierdo = Nodo("R")
raiz.derecho.derecho = Nodo("N")


def recorrer(nodo):
    if nodo:
        recorrer(nodo.izquierdo)
        print(nodo.valor, end=" ")
        recorrer(nodo.derecho)


def recorrerp(nodo):
    if nodo:
        print(nodo.valor, end=" ")
        recorrerp(nodo.izquierdo)
        recorrerp(nodo.derecho)
#Prueba:
recorrer(raiz)
```
